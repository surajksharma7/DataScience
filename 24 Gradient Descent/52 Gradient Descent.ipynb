```python
from sklearn.datasets import load_diabetes

import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test  = train_test_split(X,y,random_state = 2, test_size = 0.2)
```

```python
reg = LinearRegression()
reg.fit(X_train,y_train)
```

```python
print(reg.coef_)
print(reg.intercept_)
```

```python
y_pred = reg.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
class GDRegressor:
    def __init__(self,learning_rate=0.1,epochs=100):
        self.intercept = None
        self.coef = None
        self.epochs = epochs
        self.lr = learning_rate
    def fit(self,X_train,y_train):
        self.intercept = 0
        self.coef = np.ones(X_train.shape[1])

        for i in range(self.epochs) :
            y_hat = np.dot(X_train,self.coef) + self.intercept
            slope_intercept = -2 * np.mean(y_train - y_hat)
            self.intercept = self.intercept - self.lr* slope_intercept

            coef_der = -2 * (np.dot((y_train - y_hat),X_train))/X_train.shape[0]
            self.coef = self.coef - (self.lr * coef_der)
        print(self.intercept,self.coef)
        
    def predict(self,X_test):
        return self.intercept + (self.coef * X_test)
```

```python
gdr = GDRegressor()
```

```python
gdr.fit(X_train,y_train)
```

```python
import numpy as np

class StochasticRegressor:
    def __init__(self, learning_rate=0.1, epochs=100):
        self.intercept = None
        self.coef = None
        self.epochs = epochs
        self.lr = learning_rate

    def fit(self, X_train, y_train):
        self.intercept = 0
        self.coef = np.ones(X_train.shape[1])

        for i in range(self.epochs):
            for j in range(X_train.shape[0]):
                idx = np.random.randint(0, X_train.shape[0])
                y_hat = np.dot(X_train[idx], self.coef) + self.intercept
                error = y_train[idx] - y_hat

                # Update both inside the loop (stochastic update)
                self.intercept = self.intercept - self.lr * (-2 * error)
                self.coef = self.coef - self.lr * (-2 * error * X_train[idx])

        print(self.intercept, self.coef)

    def predict(self, X_test):
        return self.intercept + np.dot(X_test, self.coef)
```

```python
sgd = StochasticRegressor(0.01,40)
```

```python
sgd.fit(X_train,y_train)
```

```python
y_pred = sgd.predict(X_test)
```

```python
r2_score(y_test,y_pred)
```

```python
import random
class MBRegressor:
    def __init__(self,learning_rate=0.1,epochs=100,batch_size=10):
        self.intercept = None
        self.coef = None
        self.epochs = epochs
        self.lr = learning_rate
        self.batch_size = batch_size
    
    def fit(self,X_train,y_train):
        self.intercept = 0
        self.coef = np.ones(X_train.shape[1])

        for i in range(self.epochs) :
            for j in range(int(X_train.shape[0]/self.batch_size)):
                idx = random.sample(range(X_train.shape[0]),self.batch_size)
                
                y_hat = np.dot(X_train[idx],self.coef) + self.intercept
                slope_intercept = -2 * np.mean(y_train[idx] - y_hat)
                self.intercept = self.intercept - self.lr* slope_intercept

                
                coef_der = -2 * (np.dot((y_train[idx] - y_hat),X_train[idx]))/self.batch_size
                self.coef = self.coef - (self.lr * coef_der)
        print(self.intercept,self.coef)
    
    def predict(self, X_test):
        return self.intercept + np.dot(X_test, self.coef)
```
