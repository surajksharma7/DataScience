# Colored ScatterPlot

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
iris['Species'].replace({'Iris-versicolor':0,'Iris-virginica':1,'Iris-setosa':2},inplace=True)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.8)
plt.colorbar()
plt.plot()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.8)
plt.colorbar()
plt.plot()
```

### Annotations

```python
batters = pd.read_csv('batter.csv')
batters
sample_df = batters.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python

plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.axhline(130)
plt.axvline(30)
plt.axhline(140)
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

### 3D surface plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
xx.shape
```

```python
z =  xx**2 + yy**2
z.shape
```

```python
fig = plt.figure(figsize=(10,8))

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```

### Contour Plot

```python
fig, ax = plt.subplots()
p = ax.contour(xx, yy, z, cmap='viridis')
fig.colorbar(p, ax=ax)
plt.show()
```

```python
fig, ax = plt.subplots()
p = ax.contourf(xx, yy, z, cmap='viridis')
fig.colorbar(p, ax=ax)
plt.show()
```
