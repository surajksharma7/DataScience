```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('default')
```

Types of Data

### 2D plot
- Bivariate Analysis
- categorical -> numerical
- numerical ->categorical
- use case -> Time series data

```python
# plotting simple plot 

price = [48000,54000,60000,57000,47000,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
 
plt.plot(batsman['index'],batsman['V Kohli'],label='virat kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit sharma')
plt.xlabel("Season")
plt.ylabel('Runs Score')
plt.title("rohit sharma vs virat kohli career comparision")
plt.legend()
plt.show()
```

```python
# limiting axes 

price = [48000,54000,60000,57000,47000,450000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
# plt.xlim(2015,2019)
```

```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
 
plt.plot(batsman['index'],batsman['V Kohli'],label='virat kohli')
plt.plot(batsman['index'],batsman['RG Sharma'],label='Rohit sharma')
plt.xlabel("Season")
plt.ylabel('Runs Score')
plt.title("rohit sharma vs virat kohli career comparision")
plt.legend()
plt.grid()
plt.show()
```

image.png


- Bivariate Analysis

- Use Case - Finding Correlation

```python
x = np.linspace(-10,10,50)
y = 10*x+3+np.random.randint(0,100,50)
```

```python
plt.scatter(x,y)
```

```python
# plt.scatter on pandas data

df = pd.read_csv('batter.csv')
df
```

```python
df = df.head(50)
```

```python
plt.scatter(df['avg'],df['strike_rate'])
plt.title('avg and strike rate analysis')
plt.xlabel('avg')
plt.ylabel('strike_rate')
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(tips['total_bill'],tips['tip'],s=tips['size']*20)
```

```python
plt.plot(tips['total_bill'],tips['tip'],'o')
```
