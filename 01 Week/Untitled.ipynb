#### 1/1! + 2/2! + 3/3! + .....

```python
# code 
n = int(input("enter number of nth "))

result = 0
fact =  1

for i in range(1,n+1):
    fact = fact*i
    result = result + (i/ fact)

print(f"the sum of factorial is {result}")
```

#### Programming problems on Strings

```python
# find the lenght of a given string without using the len function 

len = 0

string = "hello suraj"
for i in string:
    len = len +1
print(len)
```

```python
# extract username of a given email --> nitish24singh@gmail.com = nitish24@singh

text = 'nitish24singh@gmail.com'
text.split('@')[0]
```

```python
# Count the frequency of a particular character in a provided string 

# --> hello how are you is the string --> the frequency of h in string is 2

text = "hello how are you"
text.count('h')
```

```python
# write a program to which can remove a particular character from string 

string = ''
text = "nitish"

for i in text:
    if i!= 's':
        string = string+i
```

```python
print(string)
```

```python
# to check string is palindrom or not 

text = "abac"

reverse = text[::-1]
if text == reverse:
    print("String is palindrom")
else:
    print("String is not palindrom")
```
