```python
# login program and indentation
# email -> suraj@gmail.com
# password -> 1234

email = input('enter email')
password = input('enter password')

if email == 'suraj@gmail.com' and password == '1234':
  print('Welcome')
elif email == 'suraj@gmail.com' and password != '1234':
  # tell the user
  print('Incorrect password')
  password = input('enter password again')
  if password == '1234':
    print('Welcome,finally!')
  else:
    print('beta tumse na ho paayega!')
else:
  print('Not correct')
```

### Modules 

In Python, a module is a file containing Python definitions and statements, such as functions, classes, and variables, which can be imported and used in other Python programs. Modules help organize and reuse code across different projects, improving maintainability and readability.

Types of Modules:
Built-in Modules: These come pre-installed with Python and provide various functionalities such as system operations, mathematical calculations, etc. Examples include:

math: Provides mathematical functions like square roots, trigonometry, etc.
os: Allows interaction with the operating system.
sys: Provides system-specific parameters and functions.
User-Defined Modules: These are Python files created by users, typically containing functions, classes, and variables that can be reused across different projects.

Third-Party Modules: These are external modules that you can install via package managers like pip. Examples include:

numpy: Used for numerical operations.
pandas: Used for data manipulation and analysis.
requests: Used to make HTTP requests.

```python
# number gussing game
import random 

counter = 1
num = random.randint(1,100)
num1 = int(input("guess karo "))
while num1 != num:
    if num1 > num:
        print("glat! guess low")
    else:
        print("glat! guess high")
        
    num1 = int(input("guess value"))
    counter +=1 

else:
  print('correct guess')
  print('attempts',counter)
```

### Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

```python
pop = 10000
for i in range(10,1,-1):
    pop /= 1.1
    print(int(pop))
```
