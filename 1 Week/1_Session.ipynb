### Keywords and Identifiers

```python
# Keywords in Python are reserved words that have a special meaning and purpose. 
# They are part of the syntax and cannot be used as variable names, function names, or 
# any other identifiers. Python's keywords are case-sensitive, and all keywords must be 
# written in lowercase (except for True, False, and None which are capitalized).
```

```python
# Identifiers
# You can't start with a digit
name1 = 'Nitish'
print(name1)
# You can use special chars -> _
_ = 'ntiish'
print(_)
# identiers can not be keyword
```

### Literals

```python
a = 0b1010 #Binary Literals
b = 100 #Decimal Literal 
c = 0o310 #Octal Literal
d = 0x12c #Hexadecimal Literal

#Float Literal
float_1 = 10.5 
float_2 = 1.5e2 # 1.5 * 10^2
float_3 = 1.5e-3 # 1.5 * 10^-3

#Complex Literal 
x = 3.14j

print(a, b, c, d)
print(float_1, float_2,float_3)
print(x, x.imag, x.real)
```

```python
# binary
x = 3.14j
print(x.imag)
```
