#### what is pandas
pandas is a fast, powerful flexiable and easy to use open source data analysis and manipulation tool, buit on the top of python programming langauge

#### Pandas Series

A pandas Series is like a column in a table. it is a 1-D array holding data of any type

```python
import numpy as np
import pandas as pd
```

#### Series from List

```python
country = ["india", "pakistan","USA","nepal","Srilanka"]
country = pd.Series(country)
country
```

```python
# Integer Series 
runs = [1,2,3,4,5,6,7]
pd.Series(runs)
```

```python
# custom index 

subject = ['hindi','english','math','sst']
marks = [90,80,34,24]

marks = pd.Series(marks,index=subject,name="My marks")
```

```python
marks
```

```python
marks = {
    "hindi":99,
    "English":87,
    "Maths":98,
    "Social Science": 77
}

marks = pd.Series(marks,name = "shamiya")
```

#### Series Attributes

```python
# size 
marks.size
```

```python
# dtype
marks.dtype
```

```python
# name 
marks.name
```

```python
# is_unique 
marks.is_unique
```

```python
# values
marks.values
```

```python
# index
marks.index
```
