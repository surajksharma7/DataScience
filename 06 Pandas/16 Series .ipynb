#### Pandas Series

A pandas Series is like a column in a table. it is a 1-D array holding data of any type

```python
import numpy as np
import pandas as pd
```

#### Series from List

```python
country = ["india", "pakistan","USA","nepal","Srilanka"]
country = pd.Series(country)
country
```

```python
# Integer Series 
runs = [1,2,3,4,5,6,7]
pd.Series(runs)
```

```python
# custom index 

subject = ['hindi','english','math','sst']
marks = [90,80,34,24]

marks = pd.Series(marks,index=subject,name="My marks")
```

```python
marks
```

```python
marks = {
    "hindi":99,
    "English":87,
    "Maths":98,
    "Social Science": 77
}

marks = pd.Series(marks,name = "shamiya")
```

#### Series Attributes

```python
# size 
marks.size
```

```python
# dtype
marks.dtype
```

```python
# name 
marks.name
```

```python
# is_unique 
marks.is_unique
```

```python
# values
marks.values
```

```python
# index
marks.index
```

#### Series Using read_csv

```python
# with one columns
subs = pd.read_csv("subs.csv",usecols=[0]).squeeze("columns")
subs
```

```python
#  with two columns 

vk = pd.read_csv("kohli_ipl.csv",index_col='match_no',usecols=['match_no','runs']).squeeze('columns')
vk
```

```python
movies = pd.read_csv("bollywood.csv",index_col=['movie'],usecols=['movie','lead']).squeeze('columns')
movies
```

#### Series Methods
