#### Filtering DataFrame

```python
ipl.head()
```

```python
# find all the final winner 
ipl[ipl['MatchNumber'] == "Final"][['Season',"WinningTeam"]]
```

```python
# how many superOver have finishes have occured
ipl[ipl['SuperOver'] == 'Y'].shape[0]
```

```python
# how many matches havs csv won in Kolkata
ipl["City"] == 'Kolkata'
```

```python
ipl[(ipl['City'] == 'Kolkata') & (ipl['WinningTeam'] == 'Chennai Super Kings')]
```

```python
# toss winner is match winner in percentage 
(ipl[ipl['TossWinner'] == ipl["WinningTeam"]].shape[0] / ipl.shape[0]) * 100
```

```python
# movies with rating higher than 9 and vote 10000 
movies[(movies['imdb_rating'] > 8.5) & (movies['imdb_votes'] > 10000)].shape[0]
```

```python
movies[(movies['imdb_rating'] > 7.5) & (movies['genres'].str.split('|').apply(lambda x:'Action' in x))].shape[0]
```
