##### Relational Plot

- to see the statistical relation between 2 or more variable
- Biavariate Analysis

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
# scatter Plot -> axis level funtion

sns.scatterplot(tips,x='total_bill',y='tip',hue='sex',style='time',size='size')
```

```python
# scatter Plot -> axis level funtion

sns.relplot(tips,x='total_bill',y='tip')
```

```python
# line plot
import plotly.express as px
gap = px.data.gapminder()
```

```python
india = gap[gap['country'] == 'India']
```

```python
india.sample()
```

```python
sns.relplot(india,x='year',y='lifeExp',kind='line')
```

```python
sns.lineplot(india,x='year',y='lifeExp')
```

```python
# hue ->style 

temp  = gap[gap['country'].isin(['India','China','Pakistan'])]
```

```python
temp.sample()
```

```python
sns.relplot(temp,x='year',y='lifeExp',hue='country',kind='line')
```

```python
sns.lineplot(temp,x='year',y='lifeExp',hue='country')
```

```python
sns.relplot(data=tips,x='total_bill',y='tip',kind='scatter',col='sex',row='day')
plt.show()
```

```python
sns.relplot(data=gap,x='lifeExp',y='gdpPercap',kind='scatter',col='year',col_wrap=4)
```

#### Distribution plot
- Use for univaritate analysis
- Used to find out the distribution
- Range of the observation
- Central tendency
- is the data bimodal
- Are there outliers?

##### Plots under distribution plot

### Matrix plot
- heatmap
- clustermap

```python
# Heatmap

# Plot rectangular data as a color-encoded matrix
temp_df = gap.pivot(index='country',columns='year',values='lifeExp')

# axes level function
plt.figure(figsize=(15,15))
sns.heatmap(temp_df)
```

```python
# annot
temp_df = gap[gap['continent'] == 'Europe'].pivot(index='country',columns='year',values='lifeExp')

plt.figure(figsize=(15,15))
sns.heatmap(temp_df,annot=True,linewidth=0.5, cmap='summer')
```

```python
# Clustermap

# Plot a matrix dataset as a hierarchically-clustered heatmap.

# This function requires scipy to be available.

iris = px.data.iris()
iris
```

```python
sns.clustermap(iris.iloc[:,[0,1,2,3]])
```
