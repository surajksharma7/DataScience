```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
```

```python
sns.scatterplot(data= tips,x = 'total_bill',y='tip')
```

```python
# strip plot 

sns.stripplot(data=tips,x='day',y='total_bill',jitter = .5,hue='sex')
```

```python
sns.catplot(data=tips,x='day',y='total_bill',kind='strip')
```

```python
sns.swarmplot(data=tips,x='day',y='total_bill',hue='sex')
```

### Boxplot

A boxplot is a standardized way of displaying the distribution of data based on a five number summary (minimum, first quartile [Q1], median, third quartile [Q3] and maximum). It can tell you about your outliers and what their values are. Boxplots can also tell you if your data is symmetrical, how tightly your data is grouped and if and how your data is skewed.

image.png

```python
sns.boxplot(data=tips,x = 'sex',y='total_bill')
```

```python
# using catplot 

sns.catplot(data=tips,x = 'day',y='total_bill',hue='sex',kind='box')
```

```python
sns.boxplot(data=tips,y='total_bill')
```

```python
# violin plot 

sns.violinplot(data=tips,x = 'day',y='total_bill')
```

```python
sns.catplot(data=tips,x = 'day',y='total_bill',kind='violin')
```

```python
sns.catplot(data=tips,x = 'day',y='total_bill',kind='violin',hue='sex',split='True')
```

#### Barplot

When there are multiple observations inn each categoory, it also uses bootstraping to compute  confidence interval around the estimate, which is plotted using error bar

```python
sns.barplot(data=tips,x='sex',y='total_bill',hue='sex')
```

#### PointPlot

- When There are multiple observation in each category. It alsoo uses bootstrapig to compute a confidence interval around the estimmate, which is plottinng using error bar

```python
sns.pointplot(data=tips,x='sex',y='total_bill',hue='smoker')
```

#### CountPlot 

- A special case for the bar plot is when you want to show the number of observation in each category rather than computing a statics foor a second variable. this is similar to a histogram over a categorical, rather than quntative variable


```python
sns.countplot(data=tips,x='sex',hue='day')
```

```python
sns.catplot(data=tips,x='sex',hue='total_bill',col='smoker',kind='box')
```

#### RegPlot 
- regplot
- lmplot
  
In the simplest invocation, both function draw a scatterplot to two variable, x and y, and then fir the regression model y~ x and plot the resulting regression line and a 95% confidence interval for that regression

```python
sns.regplot(tips,x='total_bill',y='tip')
```

```python
sns.lmplot(tips,x='total_bill',y='tip',hue='sex')
```

```python
# residual plot 

sns.residplot(data=tips,x='total_bill',y='tip')
```

#### A second way to plot facet plot -> facetgrid

```python
# figure level -> relplot

sns.catplot(data=tips,x='sex',y='total_bill',hue='sex',col='day',row='time',kind='violin')
```

```python
# figure level -> relplot

g = sns.FacetGrid(data=tips,hue='sex',col='day',row='time')
g.map(sns.violinplot,'sex','total_bill',)
g.add_legend()
```

```python
sns.pairplot(iris)
```

```python
sns.pairplot(iris,hue='species')
```

```python
g = sns.PairGrid(iris,hue='species')
g.map(sns.scatterplot)
g.map_offdiag(sns.rugplot)
```
