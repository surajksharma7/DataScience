#### np.sort 
return a sorted copy of an array

```python
# code 
import numpy as np
a = np.random.randint(1,100,15)
a
```

```python
np.sort(a)
```

```python
b = np.random.randint(1,100,24).reshape(4,6)
b
```

```python
np.sort(b,axis=0)
```

```python
np.sort(b,axis=1)
```
