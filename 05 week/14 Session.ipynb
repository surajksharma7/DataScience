#### Numpy array vs Python Lists

##### speed -->list

```python
a = [i for i in range(10000000)]
b = [i for i in range(10000000,20000000)]

c = []
import time

start = time.time()
for i in range(len(a)):
    c.append(a[i]+b[i])
print(time.time() - start)
```

##### speed --> Numpy

```python
import numpy as np
a = np.arange(10000000)
b = np.arange(10000000,20000000)
start = time.time()
c = a+b

print(time.time()- start)
```

```python
# total times difference 
1.541440486907959/0.12083625793457031
```

#### BoradCasting in numpy 
The term broadcasting describers how Numpy treats array with different shape during arthmetic operations. THe smalller array is broadcast across the larger so that they have much compatiable shape

```python
# same shape
a = np.arange(6).reshape(2,3)
b = np.arange(6,12).reshape(2,3)

print("matrix a is ")
print(a)
print("matrix b is ")
print(b)
print("matrix a  and b is ")
print(a+b)
```

```python
a = np.arange(12).reshape(3,4)
b = np.arange(12).reshape(4,3)
a+b
```

```python
a = np.arange(16).reshape(4,4)
b = np.arange(4).reshape(2,2)
a+b
```
