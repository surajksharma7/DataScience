```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Colored ScatterPlot

```python
iris = pd.read_csv('iris.csv')
iris.sample(5)
iris['Species'].replace({'Iris-versicolor':0,'Iris-virginica':1,'Iris-setosa':2},inplace=True)
```

```python
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.8)
plt.colorbar()
plt.plot()
```

```python
plt.figure(figsize=(15,7))
plt.scatter(iris['SepalLengthCm'],iris['PetalLengthCm'],c=iris['Species'],cmap='winter',alpha=0.8)
plt.colorbar()
plt.plot()
```

### Annotations

```python
batters = pd.read_csv('batter.csv')
batters
sample_df = batters.head(100).sample(25,random_state=5)
```

```python
sample_df
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
```

```python
plt.scatter(sample_df['avg'],sample_df['strike_rate'])
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

```python

plt.scatter(sample_df['avg'],sample_df['strike_rate'])
plt.axhline(130)
plt.axvline(30)
plt.axhline(140)
for i in range(sample_df.shape[0]):
    plt.text(sample_df['avg'].values[i],sample_df['strike_rate'].values[i],sample_df['batter'].values[i])
```

### Subplots

```python
batters.head()
```

```python
fig, ax = plt.subplots(figsize=(15,6))
ax.scatter(sample_df['avg'],sample_df['strike_rate'],color='red')
ax.set_title('avg and strike rate')
ax.set_xlabel('avg')
ax.set_ylabel('strike rate')
plt.show()
```

```python
fig , ax = plt.subplots(nrows = 2,ncols=1)
ax[0].scatter(sample_df['avg'],sample_df['strike_rate'],color='red')
ax[1].scatter(sample_df['avg'],sample_df['strike_rate'],color='red')
plt.show()
```

### 3D plots

```python
batters
```

```python
fig = plt.figure()
ax = plt.subplot(projection='3d')
ax.scatter(batters['runs'],batters['avg'],batters['strike_rate'])
ax.set_title('3d plot of scatterplot')
ax.set_xlabel('runs')
ax.set_ylabel('average')
ax.set_zlabel('strike rate')
```

### 3D surface plots

```python
x = np.linspace(-10,10,100)
y = np.linspace(-10,10,100)
```

```python
xx, yy = np.meshgrid(x,y)
xx.shape
```

```python
z =  xx**2 + yy**2
z.shape
```

```python
fig = plt.figure()

ax = plt.subplot(projection='3d')
p = ax.plot_surface(xx,yy,z,cmap='viridis')
fig.colorbar(p)
```
